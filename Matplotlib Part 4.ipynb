# Pie chart graph

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']

plt.pie(x)
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']

plt.pie(x,labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']
ex=[0.0,0.4,0.0,0.0]

plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']
ex=[0.0,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']
ex=[0.0,0.2,0.0,0.0]

plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']
c=['yellow','magenta','brown','aqua']

plt.pie(x,labels=y,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']

plt.pie(x,labels=y,autopct='%0.1f%%')
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']

plt.pie(x,labels=y,autopct='%0.2f%%',shadow=True)
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']
ex=[0.0,0.2,0.0,0.0]

plt.pie(x,labels=y,explode=ex,shadow=True,wedgeprops={'linewidth':4})
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']
ex=[0.0,0.2,0.0,0.0]

plt.pie(x,labels=y,explode=ex,shadow=True,wedgeprops={'linewidth':4,'edgecolor':'magenta'})
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']
ex=[0.0,0.2,0.0,0.0]

plt.pie(x,labels=y,explode=ex,shadow=True,rotatelabels=True)
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']

plt.pie(x,labels=y,shadow=True)
plt.title('Subject analysis')
plt.legend(loc=10)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']
x1=[5,15,25,35]
c=['yellow','magenta','red','black']

plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,labels=y,radius=0.5,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']
x1=[5,15,25,35]

plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors='w')
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Maths','Science']
x1=[5,15,25,35]

plt.pie(x,labels=y,radius=1.5)
cr=plt.Circle(xy=(0,0),radius=1,facecolor='w')
plt.gca().add_artist(cr)

plt.show()
```

# Stem plot graph

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':',markerfmt='r+')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':',markerfmt='g*')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',bottom=2)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',bottom=9)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',basefmt='g')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',basefmt='g',label='code')
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',basefmt='g',label='code',use_line_collection=False)
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',basefmt='g',label='code',orientation='horizontal')
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,2,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',basefmt='g',label='code',orientation='vertical')
plt.legend()
plt.show()
```
