```python
df=sns.load_dataset('tips').head(20)
```

```python
df
```

```python
sns.catplot(x='tip',y='size',data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.catplot(x='tip',y='size',data=df,hue='sex')
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.catplot(x='tip',y='size',data=df,hue='sex',palette='hot_r')
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.catplot(x='tip',y='size',data=df,hue='sex',palette='seismic',height=10)
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.catplot(x='tip',y='size',data=df,hue='sex',palette='seismic',height=5,kind='bar')
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.catplot(x='time',y='total_bill',data=df,hue='sex',palette='seismic',height=5,kind='box')
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.catplot(x='tip',y='size',data=df,hue='sex',palette='seismic',height=5,kind='point')
plt.xticks(rotation='vertical')
plt.show()
```

# Styling plot

```python
df
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.set_style('ticks')
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.set_context('poster',font_scale=2)
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.set_context('poster')
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
sns.despine()
plt.show()
```

# Facet grid

```python
df
```

```python
fg=sns.FacetGrid(df,col='sex',hue='day')
fg.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(df,col='sex',hue='day')
fg.map(plt.bar,'total_bill','tip').add_legend()
plt.show()
```
