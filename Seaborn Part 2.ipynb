```python
var=np.linspace(1,10,10).reshape(2,5)
```

```python
var
```

```python
ar=np.array([['a0','a1','a2','a3','a4'],['b0','b1','b2','b3','b4']])
```

```python
ar
```

```python
sns.heatmap(var,vmin=0,vmax=20,cmap='gist_heat',annot=ar,fmt='s')
plt.show()
```

```python
y=({'fontsize':20,'color':'orange'})
sns.heatmap(var,vmin=0,vmax=20,cmap='gist_heat',annot=ar,fmt='s',annot_kws=y)
plt.show()
```

```python
y=({'fontsize':20,'color':'orange'})
sns.heatmap(var,vmin=0,vmax=20,cmap='gist_heat',annot=ar,fmt='s',annot_kws=y,linewidth=4)
plt.show()
```

```python
y=({'fontsize':20,'color':'orange'})
sns.heatmap(var,vmin=0,vmax=20,cmap='gist_heat',annot=ar,fmt='s',annot_kws=y,linewidth=4,linecolor='b')
plt.show()
```

```python
y=({'fontsize':20,'color':'orange'})
sns.heatmap(var,vmin=0,vmax=20,cmap='gist_heat',annot=ar,fmt='s',annot_kws=y,linewidth=4,linecolor='b',cbar=False)
plt.show()
```

```python
y=({'fontsize':20,'color':'orange'})
sns.heatmap(var,vmin=0,vmax=20,cmap='gist_heat',annot=ar,fmt='s',annot_kws=y,linewidth=4,linecolor='b',cbar=False,xticklabels=False)
plt.show()
```

```python
y=({'fontsize':20,'color':'orange'})
sns.heatmap(var,vmin=0,vmax=20,cmap='gist_heat',annot=ar,fmt='s',annot_kws=y,linewidth=4,linecolor='b',cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
y=({'fontsize':20,'color':'orange'})
v=sns.heatmap(var,vmin=0,vmax=20,cmap='gist_heat',annot=ar,fmt='s',annot_kws=y,linewidth=4,linecolor='b',cbar=False,yticklabels=False)
v.set(xlabel='python',ylabel='program')
plt.show()
```

# Count plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
plt.show()
```

```python
sns.countplot(x='sex',data=df)
plt.show()
```

```python
sns.countplot(x='sex',data=df,hue='smoker')
plt.show()
```

```python
sns.countplot(y='sex',data=df,hue='smoker')
plt.show()
```

```python
sns.countplot(x='sex',data=df,hue='smoker',color='r')
plt.show()
```

```python
sns.countplot(x='sex',data=df,hue='smoker',palette='Dark2')
plt.show()
```

```python
sns.countplot(x='sex',data=df,hue='smoker',palette='YlGnBu')
plt.show()
```

```python
sns.countplot(x='sex',data=df,hue='smoker',palette='YlGnBu',saturation=0.2)
plt.show()
```
