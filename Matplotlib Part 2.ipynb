# scatter plot graph

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]

plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]

plt.title('Scatter plot')
plt.xlabel('month')
plt.ylabel('number')

plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
z=[10,11,12,13,14,15]

plt.title('Scatter plot',fontsize=20)
plt.xlabel('month',fontsize=20)
plt.ylabel('number',fontsize=20)

size=[300,170,90,110,130,150]
colors=[10,20,30,45,78,99]

plt.scatter(x,y,c=colors,s=size,cmap='jet')
t=plt.colorbar()
t.set_label('colorbar',fontsize=20)

plt.scatter(x,z,c=colors,s=size,cmap='Greys')
t=plt.colorbar()
t.set_label('colorbar',fontsize=20)

plt.show()
```
