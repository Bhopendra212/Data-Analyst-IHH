# Box-Whisker plot graph

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)
plt.show()
```

# Stack plot graph

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area=[1,2,3,4,5]
plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area=[2,5,3,4,1]
plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
plt.stackplot(x,area1,area2,area3)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['red','aqua','yellow'])
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['red','aqua','yellow'],baseline='sym')
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['red','aqua','yellow'],baseline='zero')
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['red','aqua','yellow'],baseline='wiggle')
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['red','aqua','yellow'])
plt.title('Graph')
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['red','aqua','yellow'])
plt.title('Graph')
plt.xlabel('X- axis')
plt.ylabel('Y-axis')
plt.legend(loc=1)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['red','aqua','yellow'])
plt.title('Graph',fontsize=20)
plt.xlabel('X- axis',fontsize=20)
plt.ylabel('Y-axis',fontsize=20)
plt.legend(loc=1)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,5,3,4,1]
area2=[1,2,3,4,2,]
area3=[6,3,5,2,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['red','aqua','yellow'])
plt.title('Graph',fontsize=20)
plt.xlabel('X- axis',fontsize=20)
plt.ylabel('Y-axis',fontsize=20)
plt.legend(loc=1)
plt.grid()
plt.show()
```
