```python
import numpy as np
import matplotlib.pyplot as plt
x=['car','bike','scooty','bicycle']
y=[40,85,75,35]
z=[20,30,40,15]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title('Parking survey')
plt.xlabel('Vehicles')
plt.ylabel('Numbers')

plt.bar(p,y,width=0.2,color='r',edgecolor='blue')
plt.bar(p1,z,width=0.2,color='g',edgecolor='blue')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=['car','bike','scooty','bicycle']
y=[40,85,75,35]
z=[20,30,40,15]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title('Parking survey')
plt.xlabel('Vehicles')
plt.ylabel('Numbers')

plt.xticks(p+width,x)

plt.bar(p,y,width=0.2,color='r',edgecolor='blue')
plt.bar(p1,z,width=0.2,color='g',edgecolor='blue')

plt.show()
```
