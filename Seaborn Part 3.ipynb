# Voilin plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
sns.violinplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue='time')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue='time',linewidth=1)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='jet_r')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='turbo_r')
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=df,order=['Dinner','Lunch'],color='r',saturation=0.5)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue='sex',split=True)
plt.show()
```

```python
sns.violinplot(y='day',x='total_bill',data=df,hue='sex',split=True)
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=df,order=['Dinner','Lunch'],inner='quartz')
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=df,order=['Dinner','Lunch'],inner='box')
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=df,order=['Dinner','Lunch'],inner='point')
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=df,order=['Dinner','Lunch'],inner='stick')
plt.show()
```

```python
sns.violinplot(x=df['total_bill'])
plt.show()
```

```python
sns.violinplot(y=df['total_bill'])
plt.show()
```

# Box plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=df,hue='time',color='g')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=df,hue='sex',color='g',order=['Fri','Sun','Thur','Sat'])
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=df,hue='sex',color='g',order=['Fri','Sun','Thur','Sat'],showmeans=True)
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=df,hue='sex',color='y',order=['Fri','Sun','Thur','Sat'],showmeans=True)
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=df,hue='sex',color='y',order=['Fri','Sun','Thur','Sat'],showmeans=True,meanprops={'marker':'+','markeredgecolor':'b'})
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=df,hue='sex',color='y',order=['Fri','Sun','Thur','Sat'],
            showmeans=True,meanprops={'marker':'+','markeredgecolor':'b'},palette='cubehelix_r')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(data=df,color='y',showmeans=True,meanprops={'marker':'+','markeredgecolor':'b'},palette='cubehelix_r')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(data=df,color='y',showmeans=True,meanprops={'marker':'+','markeredgecolor':'b'},palette='cubehelix_r',orient='h')
plt.show()
```

```python
sns.boxplot(y=df['total_bill'])
plt.show()
```
