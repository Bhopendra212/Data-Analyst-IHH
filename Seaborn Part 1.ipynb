# Line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var=[1,2,3,4,5,6,7]
var1=[7,6,5,4,3,2,1]
```

```python
plt.plot(var,var1)
plt.show()
```

```python
var=[1,2,3,4,5,6,7]
var1=[7,6,5,4,3,2,1]
import pandas as pd

df=pd.DataFrame({'var':var,'var1':var1})
sns.lineplot(x='var',y='var1',data=df)
```

# Bar Plot

```python
df1=sns.load_dataset('penguins')
```

```python
df1
```

```python
sns.barplot(x=df1.island,y=df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1)
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex')
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',order=['Dream','Torgersen','Biscoe'])
plt.show()
```

# Histogram plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=sns.load_dataset('penguins')
```

```python
df1
```

```python
sns.displot(df1['flipper_length_mm'])
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230])
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230],kde=True)
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230],kde=True,rug=True)
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230],kde=True,rug=True,color='r')
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'],color='r',log_scale=True)
plt.show()
```
