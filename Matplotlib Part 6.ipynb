# Step plot graph

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,4,1]

plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,4,1]

plt.step(x,y)
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,4,1]

plt.step(x,y)
plt.grid()
plt.title('Graph',fontsize=20)
plt.xlabel('X-axis',fontsize=20)
plt.ylabel('Y-axis',fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,4,1]

plt.step(x,y,color='r')
plt.grid()
plt.title('Graph',fontsize=20)
plt.xlabel('X-axis',fontsize=20)
plt.ylabel('Y-axis',fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,4,1]

plt.step(x,y,color='r',marker='o')
plt.grid()
plt.title('Graph',fontsize=20)
plt.xlabel('X-axis',fontsize=20)
plt.ylabel('Y-axis',fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,4,1]

plt.step(x,y,color='r',marker='o',ms=10)
plt.grid()
plt.title('Graph',fontsize=20)
plt.xlabel('X-axis',fontsize=20)
plt.ylabel('Y-axis',fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,4,1]

plt.step(x,y,color='r',marker='o',ms=15,mfc='b')
plt.grid()
plt.title('Graph',fontsize=20)
plt.xlabel('X-axis',fontsize=20)
plt.ylabel('Y-axis',fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,4,1]

plt.step(x,y,color='r',marker='o',ms=15,mfc='b',label='price')
plt.grid()
plt.title('Graph',fontsize=20)
plt.xlabel('X-axis',fontsize=20)
plt.ylabel('Y-axis',fontsize=20)
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,4,1]

plt.step(x,y,color='r',marker='o',ms=15,mfc='b',label='price')
plt.grid()
plt.title('Graph',fontsize=20)
plt.xlabel('X-axis',fontsize=20)
plt.ylabel('Y-axis',fontsize=20)
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,4,1]

plt.step(x,y,color='r',marker='o',ms=15,mfc='b',label='price')
plt.grid()
plt.title('Graph',fontsize=20)
plt.xlabel('X-axis',fontsize=20)
plt.ylabel('Y-axis',fontsize=20)
plt.legend(loc=4)
plt.show()
```

# Fill between plot graph

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y)
plt.fill_between(x,y)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color='r')
plt.fill_between(x,y)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color='r')
plt.fill_between(x,y)
plt.title('Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color='r')
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.title('Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color='r')
plt.fill_between(x,y,where=(x>=2) & (x<=4))
plt.title('Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color='r')
plt.fill_between(x,y,where=(x>=3) & (x<=4.5))
plt.title('Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color='r')
plt.fill_between(x,y,where=(x>=2) & (x<=4),alpha=0.4)
plt.title('Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

# Sub plot graph

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[1,2,3,4]
plt.plot(x,y,color='r')

x1=[1,2,3,4,5]
y1=[2,3,4,5,1]
plt.step(x,y)

x3=[10,20,30,40]
plt.pie(x3)

x4=['a','b','c','d']
y4=[2,3,4,5]
plt.bar(x4,y4)
plt.show()
```

```python
x=[1,2,3,4]
y=[1,2,3,4]

plt.subplot(2,2,1)
plt.plot(x,y,color='r')

x1=[1,2,3,4,5]
y1=[2,3,4,5,1]
plt.subplot(2,2,2)
plt.step(x,y)

x3=[10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x3)

x4=['a','b','c','d']
y4=[2,3,4,5]
plt.subplot(2,2,4)
plt.bar(x4,y4)
plt.show()
```

```python
x=[1,2,3,4]
y=[1,2,3,4]

plt.subplot(3,3,1)
plt.plot(x,y,color='r')

x1=[1,2,3,4,5]
y1=[2,3,4,5,1]
plt.subplot(3,3,2)
plt.step(x,y)

x3=[10,20,30,40]
plt.subplot(3,3,3)
plt.pie(x3)

x4=['a','b','c','d']
y4=[2,3,4,5]
plt.subplot(3,3,4)
plt.bar(x4,y4)
plt.show()
```
